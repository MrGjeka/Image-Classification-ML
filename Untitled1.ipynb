REMOVING IMAGES THAT ARE NOT SUPPORTED

```python
import cv2
from PIL import Image
```

```python
data_dir = 'Input data'   #ability to call on our input data folder to access the images
```

```python
image_exts = ['jpeg', 'jpg', 'bpm', 'png'] # allowed image types
```

```python




for image_class in os.listdir(data_dir):
    class_dir = os.path.join(data_dir, image_class)
    
    for image in os.listdir(class_dir):
        image_path = os.path.join(class_dir, image)
        try:
            with Image.open(image_path) as img_pil:
                tip = img_pil.format.lower()  # Get the image format and convert it to lowercase
            
            if tip not in image_exts:  # Check outside the 'with' statement
                print(f"Image not in ext list: {image_path}")
                os.remove(image_path)  # Remove the image file
        except Exception as e:
            print(f"Issue with image {image_path}: {e}")

          
```

SCALING THE DATA`

```python
data = data.map(lambda x,y: (x/255, y))
```

```python
data.as_numpy_iterator().next()
```

```python
batch[0].min()
```

Splitting the Data up

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
    #Class 0 = old
```

```python
len(data)
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python

train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

Deep Learning Model

```python
train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, Input
```

```python

model = Sequential()
```

```python

model = Sequential()
model.add(Input(shape=(256, 256, 3)))  # Define input shape here
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

TRAIN

```python
logdir='logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

Performance

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

Evaluation

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(pre.result(), re.result(), acc.result())
```
